```python
# Read the txt file and store word counts in a dictionary
wordCounts = {}
totalCount = 0.0

with open('hw1_word_counts_05-1.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        word = line.split(' ')[0]
        count = float(line.replace('\n', '').split(' ')[1])
        wordCounts[word] = count
        totalCount = totalCount + count

# Sort the dictionary on word counts
wordCounts = {word: count for word, count in sorted(wordCounts.items(), key = lambda element: element[1], reverse=True)}

# Compute Prior Probability dictionary
priorProbs = {word: wordCounts[word]/totalCount for word in wordCounts}
```

```python
# Question 1.9 part A
print("15 most frequent 5-letter words :-");
print(list(priorProbs.keys())[:15])

print("14 least frequent 5-letter words :-");
print(list(priorProbs.keys())[-14:])
```

```python
# Question 1.9 part B
def ProbEGivenW(evidence, word):
    # rightGuesses contains the correctly guessed letters and their indices
    rightGuesses = evidence['correct']
    
    # wrongGuesses contains the incorrectly guessed letters
    wrongGuesses = evidence['incorrect']
    
    lettersGuessedCorrectly = []
    blankPositions = [0, 1, 2, 3, 4]
    
    for guess in rightGuesses:
        lettersGuessedCorrectly.append(guess[0])
        blankPositions.remove(guess[1])
    
    prob = 1.0
    
    # If word does not contain the right guesses then prob = 0
    for guess in rightGuesses:
        if(guess[0] != word[guess[1]]):
            prob = 0.0
            break
            
    # If word contains wrong guesses then prob = 0
    for guess in wrongGuesses:
        if(guess in word):
            prob = 0.0
            break;
           
    # If letters in word corresponding to blank positions match with some correct guess then prob = 0
    for pos in blankPositions:
        if(word[pos] in lettersGuessedCorrectly):
            prob = 0.0
            break;
            
    return prob

def MostLikelyLetterAndProb(evidence):
    postProbs = {}
    postProbsDenom = 0.0
    
    # Computing posterior probability i.e. P(W=w|E)
    for word in priorProbs:
        probEGivenW = ProbEGivenW(evidence, word)
        postProbs[word] = probEGivenW * priorProbs[word]
        postProbsDenom = postProbsDenom + postProbs[word]
        
    for word in postProbs:
        postProbs[word] = postProbs[word]/postProbsDenom
    
    # Finding out guessed letters and blank positions
    rightGuesses = evidence['correct']
    wrongGuesses = evidence['incorrect']
    lettersGuessed = []
    blankPositions = [0, 1, 2, 3, 4]
    
    for guess in rightGuesses:
        lettersGuessed.append(guess[0])
        blankPositions.remove(guess[1])
        
    lettersGuessed.extend(wrongGuesses)
    
    # maxProb denotes probability of next most likely letter
    maxProb = 0.0
    mostLikelyLetter = ''
    
    for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
        if(letter not in lettersGuessed):
            prob = 0.0
            for word in postProbs:
                isLetterInWord = False
                
                # Checking if letter appears in any blank position of word
                for pos in blankPositions:
                    if(word[pos] == letter):
                        isLetterInWord = True
                        break
                    
                if(isLetterInWord):
                    prob = prob + postProbs[word]
                    
            if(maxProb < prob):
                maxProb = prob
                mostLikelyLetter = letter
                    
    return mostLikelyLetter, maxProb
```
